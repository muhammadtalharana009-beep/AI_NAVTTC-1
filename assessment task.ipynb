```python
def cal_grade(marks):
    if marks>=90:
        print('Your grade is A')
    elif (marks >=80) and (marks<=89):
        print('Your grade is B')
    elif (marks>=70) and (marks <=79):
        print('Your grade is C')
    else:
        print('You are Fail')
cal_grade(85)
```

```python
def factorial(num):
    if num <=0:
        print('Factorial is not possible. ')
    else:
        fact = 1
    for i in range(1,num+1):
        fact *= i
    return fact
factorial(5)
```

```python
def find_max(a,b,c):
    if (a >=b) and (a>=c):
        return a
    elif(b >= a) and (b>=c):
        return b
    else:
        return c
find_max(1,2,0)
```

```python
def is_prime(num):
    if num <=1:
        print('Not prime.')
    else:
        prime = True
        for i in range(2,num):
            if num%i == 0:
                prime = False
                break
        if prime:
            print('Number is prime.')
        else:
            print('Number is not prime.')
is_prime(4)
```

```python
def list_sum(num):
    sums = 0
    for i in num:
        sums += i
    return sums
list_sum([10,20,30])
```

```python
def check_pass(pas):
    if len(pas)>=8:
        is_digit = False
        for ch in pas:
            if ch.isdigit():
                is_digit = True
                
        if is_digit:
            return 'Strong'
        else:
            return 'Weak'
    else:
        return 'Weak'
check_pass('Adeelmeo3')
```

```python
def remove_dups(num):
    new_num = []
    for i in num:
        if i in new_num:
            continue
        else:
            new_num.append(i)
    return new_num
remove_dups([1,2,2,3,4,4])
```
